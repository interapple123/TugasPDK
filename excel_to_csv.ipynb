```python
import pandas as pd

# Baca file Excel
excel_file = 'raw_data/Logsheet gabungan AWS dan Pasut.xlsx'  
sheet_name = 'DUPLIKAT'     
csv_file_pasut = 'csv_data/20240714_SUROS_PRAMUKA_PASUT_15min_.csv'
csv_file_aws = 'csv_data/20240714_SUROS_PRAMUKA_AWS_15min_.csv'     

df = pd.read_excel(excel_file, sheet_name=sheet_name, converters={'Tahun': str, 'Bulan': str, 'Tanggal' : str, 'Waktu' : str})  
df
```

```python
koloms = df.columns
tes = list(koloms)
```

```python
# Temperatur
header7 = koloms[6] # Temperatur akan diubah menjadi kelembaban in
header8 = koloms[7] # Unnamed akan diubah menjadi temperatur out

# Kelembaban
header9 = koloms[8] # Kelembaban akan diubah menjadi kelembaban in
header10 = koloms[9] # Unnamed akan diubah menjadi kelembaban out

# Pasang surut
header13 = koloms[12]  # Pasang surut akan diubah menjadi pasang surut batas atas
header14 = koloms[13] # Unnamed akan diubah menjadi pasang surut batas bawah
header15 = koloms[14] # Unnamed akan diubah menjadi pasang surut rata-rata

# Memperbaiki header temperatur in and out
df.rename(columns={header7: header7 + " In"}, inplace=True)
df.rename(columns={header8: header7 + " Out"}, inplace=True)

#Memperbaiki header kelembaban in and out
df.rename(columns={header9: header9 + " In"}, inplace=True)
df.rename(columns={header10: header9 + " Out"}, inplace=True)

# Memperbaiki header pasang surut
df.rename(columns={header13: header13 + " Batas Atas (cm)"}, inplace=True)
df.rename(columns={header14: header13 + " Batas Bawah (cm)"}, inplace=True)
df.rename(columns={header15: header13 + " Rata-Rata (cm)"}, inplace=True)

df
```

```python
df = df.drop(index=0)

df = df.reset_index(drop=True)

df
```

```python
dfpasut = df.drop(df.columns[[4, 5, 6, 7, 8, 9, 10, 11]], axis=1)  
dfaws = df.drop(df.columns[[12, 13, 14]], axis=1)  
dfpasut
```

```python
# Metadata
metadata_aws = [
    "# Metadata file CSV - AWS (Automatic Weather Station)",
    "# Dibuat oleh: Firly Hafiz Syahreza",
    "# Tanggal pembuatan: 7 Januari 2025",
    "# Deskripsi: File ini berisi data pengukuran parameter cuaca dari Automatic Weather Station (AWS) di lokasi pengamatan.",
    "# Sumber data: Logsheet Excel yang telah dikonversi ke format CSV.",
    "# Lokasi: Pulau Pramuka, Kepulauan Seribu",
    "# Periode pengamatan: 14 Juli 2024 - 22 Juli 2024",
    "# Format waktu: Waktu lokal (WIB)",
    "# Penjelasan kolom:",
    "# - Tahun: Tahun pengamatan.",
    "# - Bulan: Bulan pengamatan.",
    "# - Tanggal: Tanggal pengamatan.",
    "# - Waktu: Waktu pengamatan (format HH:MM:SS).",
    "# - Keceptan Angin (m/s): Kecepatan angin dalam m/s.",
    "# - Arah angin (°): Arah angin dalam derajat.",
    "# - Temperatur (°C) In: Suhu udara disekitar konsol dalam derajat Celsius.",
    "# - Temperatur (°C) Out: Suhu udara disekitar AWS dalam derajat Celsius.",
    "# - Kelembapan (%) In: Kelembapan disekitar konsol udara dalam persen.",
    "# - Kelembapan (%) Out: Kelembapan disekitar AWS udara dalam persen.",
    "# - Curah Hujan: Curah hujan dalam mm.",
    "# - Tekanan (hpa): Tekanan udara dalam hPa (hektopascal)."
]

metadata_pasut = [
    "# Metadata file CSV - PASUT (Pasang Surut)",
    "# Dibuat oleh: Firly Hafiz Syahreza",
    "# Tanggal pembuatan: 7 Januari 2025",
    "# Deskripsi: File ini berisi data pengukuran pasang surut air laut dari lokasi pengamatan.",
    "# Sumber data: Logsheet Excel yang telah dikonversi ke format CSV.",
    "# Lokasi: Pulau Pramuka, Kepulauan Seribu",
    "# Periode pengamatan: 14 Juli 2024 - 22 Juli 2024",
    "# Format waktu: Waktu lokal (WIB)",
    "# Penjelasan kolom:",
    "# - Tahun: Tahun pengamatan.",
    "# - Bulan: Bulan pengamatan.",
    "# - Tanggal: Tanggal pengamatan.",
    "# - Waktu: Waktu pengamatan (format HH:MM:SS).",
    "# - PASANG SURUT Batas Atas (cm): Tinggi pasang surut batas atas dalam cm.",
    "# - PASANG SURUT Batas Bawah (cm): Tinggi pasang surut batas bawah dalam cm.",
    "# - PASANG SURUT Rata-Rata (cm): Rata-rata tinggi pasang surut dalam cm."
]

#AWS
# Tulis metadata dan data utama ke file CSV
with open(csv_file_aws, "w", encoding="utf-8") as f:
    for line in metadata_aws:
        f.write(line + "\n")
        
        # Menulis header kolom secara manual
    f.write(",".join(list(dfaws.columns)))
    f.write("\n")
        
    # Menulis data baris per baris
    for index, row in dfaws.iterrows():
        f.write(",".join(map(str, row.values)))
        f.write("\n")

#PASUT
# Tulis metadata dan data utama ke file CSV
with open(csv_file_pasut, "w", encoding="utf-8") as f:
    for line in metadata_pasut:
        f.write(line + "\n")
        
        # Menulis header kolom secara manual
    f.write(",".join(list(dfpasut.columns)))
    f.write("\n")
        
    # Menulis data baris per baris
    for index, row in dfpasut.iterrows():
        f.write(",".join(map(str, row.values)))
        f.write("\n")
```
